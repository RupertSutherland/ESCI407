# Kaikoura Earthquake (2016)
The NZ Community Fault Model (NZCFM) is the most recent compilation of knowledge about active faults in New Zealand (Seebeck et al. 2024; NZJGG 67: 209-229).  
Your task is to evaluate how close to failure these faults were before the earthquake.  
This notebook shows you how to load the faults.



```python
import numpy as np
import matplotlib.pyplot as plt

filename = '../data/nzcfm_kaikoura2016.csv'

# Load geometry of faults - 1 value for each element
faultname = np.genfromtxt(filename, delimiter=',', skip_header=7,
                          usecols=6,dtype='U64')
# centroid of triangle
x,y,z = np.genfromtxt(filename, delimiter=',', skip_header=7,
                          usecols=(0,1,2), unpack=True)
# orientation of fault setment
strike,dip = np.genfromtxt(filename, delimiter=',', skip_header=7,
                          usecols=(5,3), unpack=True)
# unique list of names
names = set(faultname)
```

And plot them...

```python
xmin = 1570
xmax = 1720
ymin = 5250
ymax = 5400

fig = plt.figure(figsize=(10,8))

plt.scatter(x/1e3,y/1e3,s=10,c=z,cmap='coolwarm')

plt.xlabel('NZTM X (km)')
plt.ylabel('NZTM Y (km)')
plt.colorbar(label='Depth (m)')
plt.xlim([xmin,xmax])
plt.ylim([ymin,ymax])
plt.savefig('../figs/kaikoura_map_faults.jpg',dpi=300)
```
